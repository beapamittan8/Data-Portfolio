# <center>Breast Cancer Classification Model</center>

The dataset selected consists of measurements of features of from images of a nuclei taken to determine if the patient has breast cancer. In this dataset, each of the features are computed from a digitized image of a fine needle aspirate (FNA) of a breast mass.  They describe characteristics of the cell nuclei present in the image. A few of the images can be found at http://www.cs.wisc.edu/~street/images/

The following lists the features found per sample and their respective descriptions:
    
#### Description of Fields
Attribute Information:

1) ID number
2) Diagnosis (M = malignant, B = benign)

Ten real-valued features are computed for each cell nucleus:

a) radius (mean of distances from center to points on the perimeter) \
b) texture (standard deviation of gray-scale values) \
c) perimeter \
d) area \
e) smoothness (local variation in radius lengths) \
f) compactness (perimeter^2 / area - 1.0) \
g) concavity (severity of concave portions of the contour) \
h) concave points (number of concave portions of the contour) \
i) symmetry \
j) fractal dimension ("coastline approximation" - 1) \

The mean, standard error and "worst" or largest (mean of the three
largest values) of these features were computed for each image,
resulting in 30 features. For instance, field 3 is Mean Radius, field
13 is Radius SE, field 23 is Worst Radius.

All feature values are recoded with four significant digits.

Data source: https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data/data

### Objective
The goal of this project is to build a machine learning model that predicts whether a nuclei of a cell and its respective measurements is a benign or malignant cell for breast cancer.

## Splitting the Data into Training and Testing Dataset
Machine learning models are trained to help make predictions for the future. It is necessary for us to randomly split the dataset into training and testing sets, where we use the training set to train the model, and then leverage the well-trained model to make predictions on the testing set.

```python
# Split data into training(70%) and testing data (30%) and use random_state=142
train, test = train_test_split(df, test_size=0.2, random_state=142)
print(train.shape)
print(test.shape)
```

```python
#Define dataframes of training and testing datasets
X_train = train.drop(['diagnosis'],axis=1)
y_train = train['diagnosis']

X_test = test.drop(['diagnosis'],axis=1)
y_test = test['diagnosis']
print("X_train shape: ", X_train.shape)
print("y_train shape: ", y_train.shape)
print("X_test shape: ", X_test.shape)
print("y_test shape: ", y_test.shape)
```

### Model 1: Logistic Regression
Logistic regression is a widely used and effective algorithm for binary classification tasks, where the goal is to predict a binary outcome (e.g., yes/no, 1/0, true/false) based on a set of features. It is also not sensitive to feature scaling. Whether the features are scaled or not, it can still provide meaningful predictions.

```python
#set up logistic regression model
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
# Doing predictions on train and test set
y_hat_train = model.predict(X_train)
y_hat_test = model.predict(X_test)
```

#### Logistic Regression - Evaluation

To evaluate a classification model we want to look at how many cases were correctly classified and how many were in error. In this case we have two outcomes - 0 or 1 for rating. The following metrics are used to evaluate the Logistic Regression Model: accuracy, confustion matrix, and AUC-ROC curve.

```python
# Evaluate the performance of your trained model
print("Accuracy score on training set: ", accuracy_score(y_train, y_hat_train))
print("Accuracy score on testing set: ", accuracy_score(y_test, y_hat_test))
```

```python
# Checking confusion matrix
print("Confusion matrix on test set: ")
print(confusion_matrix(y_test, y_hat_test))
```

The model correctly predicted 64 instances as negative (class 0) and 44 instances as positive (class 1).
The model incorrectly predicted 3 negative instances as positive, and 3 positive instances as negative.

```python
print("Confusion matrix on train set: ")
print(confusion_matrix(y_train, y_hat_train))
```

The model correctly predicted 280 instances as negative (class 0) and 150 instances as positive (class 1) in the training set.
The model incorrectly predicted 10 negative instances as positive, and 15 positive instances as negative.

```python
# Get predicted probabilities for the test set
y_prob = model.predict_proba(X_test)[:, 1]

# Compute ROC curve and AUC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = {:.4f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('LG - Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

An AUC-ROC of 0.9933 indicates that the model has a 99.33% chance of accurately classifying a randomly selected positive instance higher than a randomly selected negative instance. With an AUC-ROC of 0.993, the model has a very high true positive rate while keeping a very low false positive rate, making it a powerful tool for classification.



#### Recursive Feature Selection (RFE)
The initial model used all of the features but a better model can be built by omitting some variables. By using the sklearn Recursive Feature Estimation or RFE function, it removed the features one by one and evaluates the model each time. It them selects the best model for the target number of features. In this feature selection, we are targeting 10 features to be selected.

```python
# creating RFE object
lr_model = LogisticRegression()
rfe = RFE(estimator=lr_model, n_features_to_select=10)
rfe_model = rfe.fit(X_train, y_train)
print(rfe_model.support_)
```

The recursive feature selection conducted above shows the 10 most significant features. The dataframe below shows the respective 10 attributes that are considered. 

```python
#Show dataframe identify which ones are the important features
X_train.describe()
```

Based on the rfe model above, the 10 most important features are <b> radius_mean, concavity_mean, concave points_mean, radius_se, radius_worst, smoothness_worst, compactness_worst, concavity_worst, concave points_worst, symmetry_worst </b> We then perform logistic regression again with only these 10 features to see if the model will yield a higher accuracy this way. 

```python
#Define train and test dataframes
train1, test1 = train_test_split(df, test_size=0.3, random_state=142)
print(train1.shape)
print(test1.shape)
```

```python
#Define dataframes of training and testing datasets with top _ important features, dropping features that arent needed
x_train1 = train1.filter(['radius_mean','concavity_mean', 'concave points_mean', 'radius_se', 'radius_worst', 'smoothness_worst', 'compactness_worst', 'concavity_worst', 'concave points_worst', 'symmetry_worst'], axis=1)
y_train1 = train1['diagnosis']
x_test1 = test1.filter(['radius_mean','concavity_mean', 'concave points_mean', 'radius_se', 'radius_worst', 'smoothness_worst', 'compactness_worst', 'concavity_worst', 'concave points_worst', 'symmetry_worst'], axis=1)
y_test1 = test1['diagnosis']

print("x_train1 shape: ", x_train1.shape)
print("y_train1 shape: ", y_train1.shape)
print("x_test1 shape: ", x_test1.shape)
print("y_test1 shape: ", y_test1.shape)

rfe_model = rfe.fit(x_train1, y_train1)
```

```python
# doing evaluation - accuracy score of model
y_test_hat1 = rfe.predict(x_test1)
print("accuracy score on test set: ", accuracy_score(y_test1, y_test_hat1))
```

```python
# summarize all features
for i in range(x_train1.shape[1]):
    print('Column: %d, Selected %s, Rank: %.3f' % (i, rfe.support_[i], rfe.ranking_[i]))
```

```python
# to increment number of features, one at a time
acc_scores = []
for i in range(1,11):
    clf = LogisticRegression()
    rfe = RFE(estimator=clf, n_features_to_select=i)
    # training model
    rfe.fit(x_train1, y_train1)
    # predicting on test set
    y_pred = rfe.predict(x_test1)
    acc_score = accuracy_score(y_test1, y_pred)
    # print this
    print("Acc on test set using", i, "features: ", acc_score)
    # append to the list
    acc_scores.append(acc_score)
```

```python
# Estimating accuracy score on test set using RFE by using different number of features
estimator = LogisticRegression()
acc_scores = []
for i in range(1, 11):
    selector = RFE(estimator,n_features_to_select=i)
    selector = selector.fit(x_train1, y_train1)
    supp = selector.get_support()

    predicted = selector.predict(x_test1)
    acc_score = accuracy_score(y_test1, predicted)
    acc_scores.append(acc_score)

best = 1
for item in acc_scores:
    if item < acc_scores[best - 1]:
        best = acc_scores.index(item) + 1

plt.grid()
plt.xlabel('# No. of features')
plt.ylabel('Accuracy score on test set')
plt.plot(range(1, 11), acc_scores, marker = 'o', color = 'lightblue', markeredgewidth = 1 ,markeredgecolor = 'lightblue', markerfacecolor = 'None')
plt.plot(best, acc_scores[best-1], marker = 'o', markerfacecolor = 'lightblue')
```

Based on the graph above, the accuracy score shows an increasing trend as the number of features increases. With the limit set to 10 features, the accuracy stabilized at 9 and 10 features returning an accuracy score of 0.9649.

```python
# Get predicted probabilities for the test set
y_prob1 = rfe.predict_proba(x_test1)[:, 1]

# Compute ROC curve and AUC
fpr1, tpr1, thresholds1 = roc_curve(y_test1, y_prob1)
roc_auc1 = auc(fpr1, tpr1)

# Plot ROC curve
plt.figure()
plt.plot(fpr1, tpr1, color='darkorange', lw=2, label='ROC curve (area = {:.4f})'.format(roc_auc1))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('LG-RFE - Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

An AUC-ROC of 0.9969 indicates that the model has a 99.69% chance of accurately classifying a randomly selected positive instance higher than a randomly selected negative instance. With an AUC-ROC of 0.9969, the model has a very high true positive rate while keeping a very low false positive rate, making it a powerful tool for classification.

#### K-fold Cross Validation
K-fold cross-validation is a valuable technique that aids in optimizing model performance, reducing bias and variance in evaluation, and providing a more accurate estimate of how the model will perform on unseen data, making it an essential practice in machine learning model evaluation and tuning.

```python
# Create a KNN classifier model and use k-fold cross validation for testing accuracy estimation
clf1 = KNeighborsClassifier(n_neighbors = 10)
clf1.fit(X_train, y_train)
scores = cross_val_score(clf1, X_train, y_train, cv=10)
print('Accuracy for each round: \n', scores)
print('Average accuracy: %.4f +- %.4f' % (scores.mean(), scores.std()))
```

The average accuracy of approximately 0.9232 suggests that, on average, the KNN model predicts the correct class about 92.32% of the time across different data splits. The small standard deviation of 0.0327 indicates that the accuracy scores are relatively consistent across different folds.

```python
# Get predicted probabilities for the test set
y_prob4 = clf1.predict_proba(X_test)[:, 1]

# Compute ROC curve and AUC
fpr4, tpr4, thresholds4 = roc_curve(y_test, y_prob4)
roc_auc4 = auc(fpr4, tpr4)

# Plot ROC curve
plt.figure()
plt.plot(fpr4, tpr4, color='darkorange', lw=2, label='ROC curve (area = {:.4f})'.format(roc_auc4))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

#### Automatic model hyperparameter tuning
Hyperparameter tuning is a critical step in the machine learning model development process. It involves finding the best set of hyperparameters for a machine learning algorithm to optimize its performance and generalization capabilities. 

```python
# Define search space for parameters
parameter_grid = {'n_neighbors': range(1, 31)}

# Create the machine learning model
knn_clf = KNeighborsClassifier()
clf2 = GridSearchCV(knn_clf, parameter_grid, scoring='accuracy', cv=10)
clf2.fit(X_train, y_train)

# Identify the best parameter(s)
print('Best K value: ', clf2.best_params_['n_neighbors'])
print('The accuracy: ', clf2.best_score_)
```

The hyperparameter 'n_neighbors' has been determined to have the best value of 8. In the KNN algorithm, this means that the model takes into account the 8 nearest neighbours based on some distance metric (for example, Euclidean distance) while producing predictions for a new data point.

```python
# Get predicted probabilities for the test set
y_prob2 = clf2.predict_proba(X_test)[:, 1]

# Compute ROC curve and AUC
fpr2, tpr2, thresholds2 = roc_curve(y_test, y_prob2)
roc_auc2 = auc(fpr2, tpr2)

# Plot ROC curve
plt.figure()
plt.plot(fpr2, tpr2, color='darkorange', lw=2, label='ROC curve (area = {:.4f})'.format(roc_auc2))
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

An AUC-ROC of 0.9933 indicates that the model has a 99.33% chance of accurately classifying a randomly selected positive instance higher than a randomly selected negative instance. This demonstrates the model's outstanding ability to predict and separation capacity. AUC-ROC values greater than 0.9 are considered excellent, indicating a very high-quality model.
With an AUC-ROC of 0.9933, the model has a very high true positive rate while keeping a very low false positive rate, making it a powerful tool for classification.
